```python
import pandas as pd
import numpy as np
```

```python
customers = pd.read_excel("Northwind_Project_full.xlsx", sheet_name=0)
category = pd.read_excel("Northwind_Project_full.xlsx", sheet_name=1)
employees = pd.read_excel("Northwind_Project_full.xlsx", sheet_name=2)
employee_territories = pd.read_excel("Northwind_Project_full.xlsx", sheet_name=3)
order_details = pd.read_excel("Northwind_Project_full.xlsx", sheet_name=4)
orders = pd.read_excel("Northwind_Project_full.xlsx", sheet_name=5)
shippers = pd.read_excel("Northwind_Project_full.xlsx", sheet_name=6)
suppliers = pd.read_excel("Northwind_Project_full.xlsx", sheet_name=7)
products = pd.read_excel("Northwind_Project_full.xlsx", sheet_name=8)
regions = pd.read_excel("Northwind_Project_full.xlsx", sheet_name=9)
territories = pd.read_excel("Northwind_Project_full.xlsx", sheet_name=10)
```

```python
products = products[:77]
products["Discontinued"] = products["Discontinued"].astype(int)
products["CategoryID"] = products["CategoryID"].astype(int)
products["TotalUnits"] = products["UnitsOnOrder"] + products["UnitsInStock"]
products["TotalCost"] = products["TotalUnits"] * products["UnitPrice"]
products_name = products[["CategoryID", "ProductName", "UnitPrice", "TotalUnits", "TotalCost", "Discontinued"]]
products_name
```

```python
orders_and_details = pd.merge(order_details, orders)
orders_and_details.head()
```

```python
details_name = pd.merge(order_details, products[["ProductID", "ProductName"]], on="ProductID", how="left")
details_name["TotalPrice"] = (details_name["UnitPrice"] * details_name["Quantity"]) * (1 - details_name["Discount"])
details_name
```

```python
group_details_name = details_name.groupby(["ProductName"])[["TotalPrice"]].sum().reset_index()
group_details_name
```

```python
profits = pd.merge(group_details_name, products_name)
profits["Profit"] = profits["TotalPrice"] - profits["TotalCost"]
profits
```

```python
profit_category = pd.merge(profits, category[["CategoryID", "CategoryName"]])
profit_category
```

```python
profit_category.to_excel('profit_category.xlsx', index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

le = LabelEncoder()
profit_category['Category_Encoded'] = le.fit_transform(profit_category['CategoryName'])

features = ['UnitPrice', 'TotalUnits', 'TotalCost', 'Category_Encoded']
X = profit_category[features]
y = profit_category['Discontinued']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier(n_estimators=1)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print(classification_report(y_test, y_pred))
```
